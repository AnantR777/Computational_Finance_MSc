# Histograms

```python
#This Python script is designed to demonstrate the creation of histograms and bar charts for a set of uniformly distributed random samples. 
#Specifically, it covers three key visualizations:

# 1. Basic Histogram: It first generates a histogram of the random samples with specified bin edges. 
#    This histogram shows the frequency distribution of the samples.

# 2. Normalized Histogram with PDF Overlay: The script then creates a normalized histogram, where the heights of the bars represent 
#    probability density instead of just counts. This histogram is overlaid with the probability density function (PDF) of the uniform 
#    distribution, providing a visual comparison between the empirical distribution of the samples and the theoretical uniform distribution.

# 3. Bar Chart Representation: Lastly, the script converts the histogram data into a bar chart. This is achieved by first calculating the 
#    histogram data using np.histogram with density=True (to normalize it), and then plotting it as a bar chart. The uniform distribution's 
#    PDF is also overlaid for comparison.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Generate random samples
nsample = 50
X = np.random.rand(nsample)  # Uniformly distributed samples between 0 and 1

# Define bin edges and calculate bin width
x = np.arange(0, 1.05, 0.05)  # Bin edges from 0 to 1 in steps of 0.05
dx = x[1] - x[0]  # Bin width

# Calculate bin centers for plotting
xx = x[:-1] + dx/2

# Histogram: Basic histogram with bin edges
plt.figure(1)
plt.hist(X, bins=x, edgecolor='black', alpha=0.6)
plt.title("Basic Histogram with Bin Edges")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Histogram: Normalized histogram
plt.figure(2)
plt.hist(X, bins=x, density=True, alpha=0.6, edgecolor='black')
plt.plot(xx, uniform.pdf(xx), 'b', label="Uniform PDF")
plt.title("Normalized Histogram with Uniform PDF Overlay")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()

# Bar Chart: Representing normalized histogram data
plt.figure(3)
counts, _ = np.histogram(X, bins=x, density=True)
plt.bar(xx, counts, width=dx, alpha=0.6, edgecolor='black')
plt.plot(xx, uniform.pdf(xx), 'b', label="Uniform PDF")
plt.title("Bar Chart Representation with Uniform PDF Overlay")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()

plt.show()
```
